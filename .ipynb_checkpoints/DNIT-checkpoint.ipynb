```python
#Importação dos métodos:
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from io import StringIO
import time
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Selecionando e abrindo o Navegador:
def driversetup(isChrome):
  if(isChrome):
    print('Chrome')
    options = webdriver.ChromeOptions()
    # run Selenium in headless mode
    # options.add_argument('--headless') #Tirar o comentario caso queira que fique oculta a janela do Chrome
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
  else:
    print('Firefox')
    option = Options()
    option.headless = True
    driver = webdriver.Firefox(options=option)
  return driver
```

```python
def pagesource(url, driver):
    driver = driver
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    driver.close()
    return soup
```

```python
isChrome = True
url = 'https://servicos.dnit.gov.br/dadospnct/ContagemContinua'
driver = driversetup(isChrome)
print(url)
```

```python
import os
for ano in range(2014, 2023):
    filename = f'table{ano}.csv'
    if not os.path.exists(filename):
        print(f"Arquivo {filename} não encontrado. Pulando para o próximo ano.")
        continue  # Pula para o próximo ano se o arquivo não existe
    df = pd.read_csv(filename, sep=';')
    #df.info()
    #df.duplicated().sum()
    classeA = df[df['Classe'] == 'A']
    print(f"Gerando gráfico para o arquivo {filename}")
    sns.catplot(data=classeA, kind="bar", x="Sentido", y="total") 
    
```

#### Classe A

Ônibus/Cam de 2 eixos

```python
classeA.shape
```

```python
subA = classeA.iloc[:2, 3:35]
```

```python
C = subA[subA['Sentido'] == 'C']
C = C.drop(['Sentido'],axis =1)
C = np.array(C)
```

```python
print(f'Media de Ônibus/Cam de 2 eixos no Sentido C = {C.mean()}')
print(f'Maximo = {C.max()}')
print(f'Minimo = {C.min()}')
print(f'Desvio Padrao = {C.std()}')
print(f'Mediana = {np.median(C)}')
```

```python
D = subA[subA['Sentido'] == 'D']
D = D.drop(['Sentido'],axis =1)
D = np.array(D)
```

```python
print(f'Media de Ônibus/Cam de 2 eixos no Sentido D = {D.mean()}')
print(f'Maximo = {D.max()}')
print(f'Minimo = {D.min()}')
print(f'Desvio Padrao = {D.std()}')
print(f'Mediana = {np.median(D)}')
```

```python
plt.figure(figsize=(10, 6))
for sentido in ['C', 'D']:
    dados_sentido = subA[subA['Sentido'] == sentido]
    plt.plot(dados_sentido.columns[1:], dados_sentido.values[0][1:], label=f'Sentido {sentido}', marker='o', linestyle='-')


plt.title('Gráfico de Linha de Ônibus/Cam de 2 eixos por Sentido')
plt.xlabel('Tempo')
plt.ylabel('Valores')
plt.legend()


plt.grid(True)
plt.show()
```

```python
C = subtotal[subtotal['Sentido'] == 'C']
C = C.drop(['Sentido'],axis =1)
C = np.array(C)
print(f'Media de Veiculos no Sentido C = {C.mean()}')
print(f'Maximo = {C.max()}')
print(f'Minimo = {C.min()}')
print(f'Desvio Padrao = {C.std()}')
print(f'Mediana = {np.median(C)}')
```

```python
D = subtotal[subtotal['Sentido'] == 'D']
D = D.drop(['Sentido'],axis =1)
D = np.array(D)
print(f'Media de Motos no Sentido D = {D.mean()}')
print(f'Maximo = {D.max()}')
print(f'Minimo = {D.min()}')
print(f'Desvio Padrao = {D.std()}')
print(f'Mediana = {np.median(D)}')
```

```python
totalC = C.sum()
totalD = D.sum()
total = totalD + totalC

print(f'{((totalC/total)*100):.2f}% dos veiculos estavam no sentido C')
print(f'{((totalD/total)*100):.2f}% dos veiculos estavam no sentido D')
```

```python
plt.figure(figsize=(10, 6))
for sentido in ['C', 'D']:
    dados_sentido = subtotal[subtotal['Sentido'] == sentido]
    plt.plot(dados_sentido.columns[1:], dados_sentido.values[0][1:], label=f'Sentido {sentido}', marker='o', linestyle='-')


plt.title('Gráfico de Linha do Total de Veiculos por Sentido')
plt.xlabel('Tempo')
plt.ylabel('Valores')
plt.legend()


plt.grid(True)
plt.show()
```
